```python
# Import necessary libraries
import scrapy
import pandas as pd
from scrapy.crawler import CrawlerProcess
from scrapy.utils.project import get_project_settings

# Define the Spider class for web scraping
class CompanySpider(scrapy.Spider):
    name = "company_spider"
    start_url = "https://vdiv-bw.de/verwaltersuche/?no_cache=1&tx_browser_pi1%5Bradius%5D=50&page=1"
    allowed_domains = ["vdiv-bw.de"]

    # Start the scraping process
    def start_requests(self):
        yield scrapy.Request(url=self.start_url, callback=self.parse)

    # Parse the main page to extract links to individual company details
    def parse(self, response):
        for link in response.xpath('//a[@class="linktorecord"]'):
            yield response.follow(link, self.parse_company_details)

        # Check if there's a "Next" button/link and follow it if available
        next_page_link = response.xpath('//*[@id="c1001-tx-browser-pi1"]/div/div[3]/div[3]/div/nav/ul/li[13]/a/@href').get()
        if next_page_link:
            next_page_absolute_url = response.urljoin(next_page_link)
            yield scrapy.Request(url=next_page_absolute_url, callback=self.parse)
            
     # Parse individual company details
    def parse_company_details(self, response):
        company_info = {
            'Company Name': response.xpath('//h1/text()').get(default='').strip(),
            'Street': response.xpath('//ul[@class="vcard tx_org_headquarters address"]/li[2]/text()').get(default='').strip(),
            'Postcode and City': response.xpath('//ul[@class="vcard tx_org_headquarters address"]/li[3]/text()').get(default='').strip(),
            'Telephone': response.xpath('//ul[@class="vcard tx_org_headquarters contact"]/li[3]/text()').get(default='').replace('Tel.: ', '').strip(),
            'Email': response.xpath('//ul[@class="vcard tx_org_headquarters contact"]/li[4]//a/@href').get(default='').replace('mailto:', '').strip(),
        }
        yield company_info

# Define a custom pipeline to save scraped data to a CSV file
class SaveToCsvPipeline:
    def __init__(self):
        self.items = []

    def process_item(self, item, spider):
        self.items.append(item)
        return item

    def close_spider(self, spider):
        df = pd.DataFrame(self.items)
        df.to_csv("companies_output.csv", index=False, encoding='utf-8')
        
# Define the settings for the web crawler
settings = {
    'USER_AGENT': 'Mozilla/5.0',
    'BOT_NAME': 'scrapybot',
    'DOWNLOAD_DELAY': 3  # Add a delay of 3 seconds between requests.
}

# Create a CrawlerProcess with the loaded settings
settings = get_project_settings()
process = CrawlerProcess(settings)

# Attach the custom pipeline
process.settings.set('ITEM_PIPELINES', {'__main__.SaveToCsvPipeline': 100})

# Start the spider
process.crawl(CompanySpider)
process.start()

# Print a completion message after the scraping is done and saved to CSV
print("Scraping and saving to CSV completed!")

# Read the CSV file using pandas
csv_file = "companies_output.csv"
df = pd.read_csv(csv_file)

# Define the output Excel file path
output_file = r"C:\Users\USER\Desktop\Data Extraction\Succesful Scraping\companies_output.xlsx"

# Save the DataFrame to Excel
df.to_excel(output_file, index=False)

print("CSV file converted and saved to Excel successfully!")
```
